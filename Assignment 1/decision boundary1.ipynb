```python

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import accuracy_score

# Data
# This is the dataset provided in the assignment.
# Each row represents a sample with three columns: age, temperature, and pass/fail status (1 for pass, 0 for fail)
data = np.array([
    [15.5, 40, 0],
    [23.75, 23.25, 0],
    [8, 17, 1],
    [17, 21, 0],
    [5.5, 10, 1],
    [19, 12, 1],
    [24, 20, 0],
    [2.5, 12, 1],
    [7.5, 15, 1],
    [11, 26, 0]
])

# Features and labels
# Separating the features (age and temperature) and the labels (pass/fail status)
X = data[:, :2]
y = data[:, 2]

# kNN with k=5
# Creating an instance of KNeighborsClassifier with k=5
k = 5
knn = KNeighborsClassifier(n_neighbors=k)
# Fitting the kNN model with the data
knn.fit(X, y)

# Plot decision boundary
# Define the step size for the mesh grid
h = .02  
# Define the range for the mesh grid based on the data
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
# Create a mesh grid
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Predict the class for each point in the mesh grid
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
# Reshape the result to match the shape of the mesh grid
Z = Z.reshape(xx.shape)

# Plot the decision boundary and the data points
plt.figure()
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.Paired)
plt.xlabel('Propellant age (Weeks)')
plt.ylabel('Storage temperature (Â°C)')
plt.title(f'Decision Boundary with k = {k}')
plt.show()

# Leave-One-Out Cross-Validation (LOOCV)
# Create an instance of LeaveOneOut cross-validator
loo = LeaveOneOut()
accuracy = []  # List to store the accuracy for each iteration

# Iterate through each data point
for train_index, test_index in loo.split(X):
    # Split the data into training and test sets
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    # Fit the kNN model on the training set
    knn.fit(X_train, y_train)
    # Predict the class for the test set
    y_pred = knn.predict(X_test)
    # Calculate the accuracy and store it in the list
    accuracy.append(accuracy_score(y_test, y_pred))

# Calculate the average accuracy
average_accuracy = np.mean(accuracy)
print(f'LOOCV Accuracy with k = {k}: {average_accuracy:.2f}')

# Find the optimum k
# List to store the accuracies for different values of k
accuracies = []
# Range of k values to test
k_values = range(1, 10)

# Iterate through each value of k
for k in k_values:
    # Create an instance of KNeighborsClassifier with the current value of k
    knn = KNeighborsClassifier(n_neighbors=k)
    accuracy = []  # List to store the accuracy for each iteration
    # Perform LOOCV for the current value of k
    for train_index, test_index in loo.split(X):
        # Split the data into training and test sets
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        # Fit the kNN model on the training set
        knn.fit(X_train, y_train)
        # Predict the class for the test set
        y_pred = knn.predict(X_test)
        # Calculate the accuracy and store it in the list
        accuracy.append(accuracy_score(y_test, y_pred))
    # Calculate the average accuracy for the current value of k
    accuracies.append(np.mean(accuracy))

# Find the value of k that gives the highest accuracy
optimal_k = k_values[np.argmax(accuracies)]
print(f'Optimal k: {optimal_k}')

# Plot the LOOCV accuracy for different values of k
plt.figure()
plt.plot(k_values, accuracies, marker='o')
plt.xlabel('k')
plt.ylabel('LOOCV Accuracy')
plt.title('LOOCV Accuracy for different k values')
plt.show()
```
